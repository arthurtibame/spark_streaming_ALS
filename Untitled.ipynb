```python
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit,CrossValidator
from pyspark.ml.tuning import TrainValidationSplitModel
from pyspark.ml.recommendation import ALSModel
from pyspark.ml.feature import StringIndexer
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
df = spark.read.csv("hdfs://master/whiskey/user_whisky_score.csv", header=True)
df = df.withColumn("rating", df["score"].cast("float")).drop(df["score"])# score (str -> float)

whiskyIndexer = StringIndexer(inputCol="whisky_name", outputCol="whiskeyId",handleInvalid='error') 
userIndexer = StringIndexer(inputCol='user_name',outputCol='userId',handleInvalid='error') 

whiskyIndexed = whiskyIndexer.fit(df).transform(df) # apply whisky_name indexer
userIndexed = userIndexer.fit(whiskyIndexed).transform(whiskyIndexed) # apply user indexer

df_indexed = userIndexed.drop('whisky_name').drop('user_name') # remove old columns with 
df_indexed.show()
```

```python
from pyspark.sql.functions import lit

whisky_name = df.select(df['whisky_name']).distinct()
whisky_name = [str(row.whisky_name) for row in whisky_name.collect()]
```

```python
whisky_name = userIndexed.select(userIndexed['whisky_name']).distinct().orderBy(userIndexed.whiskeyId.asc())
whisky_name = [str(row.whisky_name) for row in whisky_name.collect()]
```

```python
user_name = userIndexed.select(userIndexed['user_name']).distinct().orderBy(userIndexed.userId.asc())
user_name = [str(row.user_name) for row in user_name.collect()]
```

```python
# split training , test set here 
(df_train, df_test) = df_indexed.randomSplit([0.7, 0.3])
```

```python
als = ALS(maxIter=5,userCol="userId",itemCol="whiskeyId",ratingCol="rating" , coldStartStrategy="drop")
#model = als.fit(df_train)

# Transform the test data using the model to get predictions
# predicted_test_data = model.transform(df_train)

# Evalute model performance with test set
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="rating", metricName="rmse")
param_grid = ParamGridBuilder().addGrid(als.rank, [12, 13, 14]).addGrid(als.maxIter, [18, 19, 20]).addGrid(als.regParam, [.17, .18, .19]).build()

evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")

tvs = TrainValidationSplit(estimator=als, estimatorParamMaps=param_grid, evaluator=evaluator)
model = tvs.fit(df_train)

# Specify the number of whisky you would like to recommand for each user
user_whiskey = model.bestModel.recommendForAllUsers(5)
user_whiskey.show(100, truncate=False)

# The users who are most likely to like a particular whisky
whiskey_users = model.bestModel.recommendForAllItems(5)
whiskey_users.show(100, truncate=False)
```
